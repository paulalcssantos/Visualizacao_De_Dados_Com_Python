#**Aula 1 - Primeiros Gráficos, Figuras, Titulos e Rótulos**

##**0. DADOS SIMULADOS**

Gerando um Dataset Simulado para Mulheres em Dados em 2025

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
n_linhas = 500
dados = {
    'Cargo': np.random.choice(['Cientista de Dados', 'Analista de Dados', 'Engenheira de Dados', 'Analista de BI'], n_linhas),
    'Nivel': np.random.choice(['Júnior', 'Pleno', 'Sênior'], n_linhas, p=[0.4, 0.4, 0.2]),
    'Salario': np.random.normal(loc=8000, scale=2000, size=n_linhas),
    'Anos_Experiencia': np.random.randint(1, 15, n_linhas),
    'Linguagem_Principal': np.random.choice(['Python', 'SQL', 'R', 'Scala'], n_linhas)
}
df = pd.DataFrame(dados)
df.loc[df['Nivel'] == 'Júnior', 'Salario'] *= 0.7
df.loc[df['Nivel'] == 'Sênior', 'Salario'] *= 1.5
df['Salario'] = df['Salario'].round(2)
display(df.head(10))
```

## **1. GRÁFICO DE LINHA (Tendência Simples)**

Vamos simular um crescimento de vagas ao longo dos últimos 5 anos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

anos = ['2021', '2022', '2023', '2024', '2025']
vagas = [150, 300, 450, 1000, 1200]

plt.figure(figsize=(8, 4))

plt.plot(anos, vagas, marker='o', linestyle='-', color='teal')

plt.title('Vagas de Cientista de Dados ao Longo dos Últimos 5 Anos')
plt.xlabel('Ano')
plt.ylabel('Número de Vagas')

plt.grid(True)

plt.show()
```

## **2. GRÁFICO DE BARRAS VERTICAIS**

```python
contagem_nivel = df['Nivel'].value_counts()

plt.figure(figsize=(8, 4))

plt.bar(contagem_nivel.index, contagem_nivel.values, color='coral', edgecolor='black')

plt.title('Contagem de Níveis de Experiência')
plt.xlabel('Nível de Experiência')
plt.ylabel('Total de Mulheres')

plt.show()
```

## **3. GRÁFICO DE BARRAS HORIZONTAIS**

```python
media_salarial_por_cargo = df.groupby('Cargo')['Salario'].mean().sort_values()

plt.figure(figsize=(10, 6))

plt.barh(media_salarial_por_cargo.index, media_salarial_por_cargo.values, color='coral', edgecolor='black')

plt.title('Média Salarial por Cargo')
plt.xlabel('Média Salarial')
plt.ylabel('Cargo')

plt.show()
```

# **Aula 2 - Subplots e Eixos Duplos**

##**Subplots**

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

'''Área Superior Esquerda: axes[0, 0]'''
# Gráfico de barras dos cargos
cargos = df['Cargo'].value_counts()

axes[0, 0].bar(cargos.index, cargos.values, color='cornflowerblue', edgecolor='black')
axes[0, 0].set_title('Contagem de Cargos')
axes[0, 0].set_xlabel('Cargo')

axes[0, 0].tick_params(axis='x', rotation=25)

'''Área Superior Direita: axes[0, 1]'''
# Gráfico de barras horizontais da média salarial

axes[0, 1].barh(media_salarial_por_cargo.index, media_salarial_por_cargo.values, color='lightpink', edgecolor='black')
axes[0, 1].set_title('Média Salarial por Cargo')
axes[0, 1].set_xlabel('Média Salarial')

axes[0, 1].tick_params(axis='x', rotation=25)


'''Área Inferior Esquerda: axes[1,0]'''
# Scatter plot relacionando experiência e salário
# (diagrama de dispersão, é uma representação visual que exibe a relação entre duas variáveis numéricas)

axes[1, 0].scatter(df['Anos_Experiencia'], df['Salario'], alpha=0.6, color='teal')
axes[1, 0].set_title('Correlação: Experiência vs Salário', fontsize=14)
axes[1, 0].set_xlabel('Anos de Experiência')
axes[1, 0].set_ylabel('Salário (R$)')

'''Área Inferior Direita: axes[1,1]'''
# Histograma simples
# (histograma é um gráfico que representa a distribuição de frequências de dados numéricos contínuos)

axes[1,1].hist(df['Salario'], bins=20, color='lightcoral', edgecolor='black')
axes[1,1].set_title('Distribuição de Salários')
axes[1,1].set_xlabel('Salário')


#Área geral (figura)
fig.suptitle('Análise de Mulheres em Dados', fontsize=16, fontweight='bold')

plt.tight_layout()

plt.show()
```

## **Eixo Secundário**

```python
# Contagem de pessoas e da Média Salarial por Nível
resumo = df.groupby('Nivel').agg(
    Total_Pessoas=('Cargo', 'count'),
    Media_Salarial=('Salario', 'mean')
).reindex(['Júnior', 'Pleno', 'Sênior']) # Força a ordem lógica de senioridade

# Figura
fig, ax1 = plt.subplots(figsize=(10, 6))

# Eixo Primário (ax1)
ax1.bar(resumo.index, resumo['Total_Pessoas'], color='lightgray')

ax1.set_ylabel('Total de Profissionais', color='gray', fontsize=12, fontweight='bold')
ax1.tick_params(axis='y', labelcolor='gray')

# Eixo Secundário (ax2)
ax2 = ax1.twinx() #cria um clone de ax1, compartilhando o eixo X, mas com um novo eixo Y na direita.

ax2.plot(resumo.index, resumo['Media_Salarial'], color='crimson', marker='o', linewidth=3)

ax2.set_ylabel('Média Salarial (R$)', color='crimson', fontsize=12, fontweight='bold')
ax2.tick_params(axis='y', labelcolor='crimson')

# Apresentação
plt.title('Análise Dupla: Volume de Pessoas vs. Remuneração Média', fontsize=15)
plt.show()
```

# **Aula 3 - Design, Cores e Exportação**

## **Paletas Customizadas e Spines**

```python
''' Definindo as cores institucionais do nosso projeto '''
# Azul escuro, rosa, Verde esmeralda, Cinza chumbo
cores_da_marca = ['#2C3E50', '#E74783', '#27AE60']

cargos = df['Cargo'].value_counts()

fig, axes = plt.subplots(figsize=(10, 6))

# Aqui, passamos nossa LISTA de cores. O Matplotlib vai iterar: 1ª barra = 1ª cor...
# Se houver mais barras que cores, ele repete a lista desde o início.
axes.bar(cargos.index, cargos.values, color=cores_da_marca)

axes.set_title('Adoção de Cargos na Empresa (Cores Institucionais)', fontsize=15)
axes.set_ylabel('Total de Contratações')
axes.tick_params(axis='x', rotation=15)

#Bordas/spines (limpeza)
axes.spines['top'].set_visible(False)
axes.spines['right'].set_visible(False)

axes.spines['left'].set_color('lightgray')
axes.spines['bottom'].set_color('lightgray')

plt.show()
```

## **Colormaps (Cores dinâmicas) e Exportação**

```python
fig, axes = plt.subplots(figsize=(12, 7))

grafico = axes.scatter(
    df['Anos_Experiencia'],
    df['Salario'],
    c=df['Anos_Experiencia'], # A coluna que dita a COR (é o nosso gradiente)
    cmap='plasma', # (O nome do Colormap oficial do Matplotlib. 'plasma' e 'viridis' são ótimos e inclusivos.)
    s=60, # O tamanho da bolinha
    alpha=0.7 # Transparência
)

axes.set_title('Mapa de Dispersão: Colormap Plasma', fontsize=16)
axes.set_xlabel('Anos de Experiência')
axes.set_ylabel('Salário Anual (R$)')
axes.grid(True, linestyle='--', alpha=0.3) # Grade bem suave

#Exportação

nome_arquivo = 'relatorio_dispersao_salarial.png'

plt.savefig(nome_arquivo, dpi=300, bbox_inches='tight')

print(f"✅ SUCESSO! A imagem '{nome_arquivo}' foi gerada com alta qualidade.")
print("Verifique o ícone de pasta (Arquivos) na barra lateral esquerda do Colab para baixar a imagem.")

plt.show()
```

# **Aula 4 - Introdução ao Seaborn**

## **COUNTPLOT (gráfico de barras automático)**

```python
import seaborn as sns

# Configurando o estilo global
sns.set_theme(style="whitegrid", palette="Set2")

sns.countplot(data=df, x='Linguagem_Principal', hue='Nivel')
plt.title('Linguagem Segmentada por Senioridade')
plt.ylabel('Total de Mulheres')

plt.show()
```

## **BARPLOT (Com agregação estatística)**

```python
plt.figure(figsize=(12, 6))

sns.barplot(data=df, x='Cargo', y='Salario', hue='Nivel', palette='viridis', errorbar=('ci', 95)) #mostra o intervalo de confiança de 95% (o tracinho preto)
plt.title('Média Salarial com Intervalo de Confiança por Cargo e Nível')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # Move a legenda para fora do gráfico
plt.tight_layout()
plt.show()
```

## **SCATTERPLOT (Com e sem linha de regressão)**

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Apenas a dispersão
sns.scatterplot(data=df, x='Anos_Experiencia', y='Salario', hue='Cargo', ax=ax1, alpha=0.7)
ax1.set_title('Relação Pura: Experiência vs Salário')

# LMPlot (Regressão Linear)
# LMPlot cria sua própria figura, então usamos regplot no subplot
sns.regplot(data=df, x='Anos_Experiencia', y='Salario', ax=ax2, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
ax2.set_title('Regressão Linear: A Tendência Central')

plt.show()
```

# **Aula 5 - Distribuições e Matrizes**

## **HISTPLOT (Histograma)**

```python
fig, axes = plt.subplots(figsize=(10, 5))

sns.histplot(data=df, x='Salario', bins=20, kde=True, color='#E74783')

plt.title('Concentração Geral de Salários', fontsize=14, fontweight='bold')
plt.xlabel('Faixa Salarial (R$)')
plt.ylabel('Quantidade de Pessoas')

axes.spines['top'].set_visible(False)
axes.spines['right'].set_visible(False)
axes.spines['left'].set_color('lightgray')
axes.spines['bottom'].set_color('lightgray')

axes.grid(False)

plt.show()
```

## **BOXPLOT vs VIOLINPLOT**

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Boxplot clássico
sns.boxplot(data=df, x='Cargo', y='Salario', hue='Cargo', ax=ax1, palette='Set3')
ax1.set_title('Boxplot: Foco em Outliers e Quartis')
ax1.tick_params(axis='x', rotation=30)

# Violinplot: O interior mostra a miniatura do boxplot, as bordas mostram a densidade (volume de pessoas naquele valor)
sns.violinplot(data=df, x='Cargo', y='Salario', hue='Cargo', ax=ax2, palette='Set3', inner="quartile")
ax2.set_title('Violinplot: Foco na Densidade da Distribuição')
ax2.tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

## **HEATMAP (Mapa de Calor)**

```python
# Correlação só funciona com dados numéricos. Precisamos filtrar o DataFrame.
df_numerico = df[['Salario', 'Anos_Experiencia']]

# Calculando a matriz de correlação de Pearson
matriz_corr = df_numerico.corr()

plt.figure(figsize=(6, 5))

sns.heatmap(matriz_corr, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)

plt.title('Matriz de Correlação')
plt.show()
```

### **Heatmap Iris**

Carregando um dataset mais complexo diretamente da biblioteca Seaborn.

Este dataset contém medidas do comprimento e largura de pétalas e sépalas de flores.


```python
df_flores = sns.load_dataset("iris")

# Mostrando as primeiras linhas para entendermos os dados
print("Amostra dos dados das Flores Iris:")
display(df_flores.head())
```

```python
# Precisamos selecionar APENAS as colunas numéricas, ignorando a coluna de texto 'species'.
colunas_numericas = df_flores.select_dtypes(include=['float64', 'int64'])

# Calculamos a matriz de correlação cruzando todas essas variáveis numéricas de uma vez
matriz_complexa = colunas_numericas.corr()


plt.figure(figsize=(8, 6))

mapa = sns.heatmap(matriz_complexa, annot=True, cmap='magma', fmt=".2f", vmin=-1, vmax=1, square=True, linewidths=0.5)

plt.title('Medidas das Flores Iris', fontsize=16, fontweight='bold',)
plt.xticks(rotation=45, ha='right')

plt.show()
```

# **Aula 6 - Gráficos Interativos com Plotly**

## **Scatter plot interativo**

```python
import plotly.express as px

fig_scatter = px.scatter(
    df,
    x='Anos_Experiencia',
    y='Salario',
    color='Cargo',          # As cores definem os cargos
    size='Salario',         # O tamanho da bolha aumenta com o salário
    hover_name='Nivel',     # O título da caixinha (tooltip) será o Nível
    hover_data=['Linguagem_Principal'], # Adiciona a linguagem na caixinha
    title='Exploração Interativa: Salário vs. Experiência',
    template='plotly_white'
)

fig_scatter.show()
```

## **TREEMAP (Mapa de Árvore)**

Uma alternativa muito mais moderna e legível que o Gráfico de Pizza.

O treemap mostra proporções através da área de retângulos aninhados.

```python
df_agrupado = df.groupby(['Cargo', 'Nivel']).size().reset_index(name='Quantidade')

fig_tree = px.treemap(
    df_agrupado,
    path=[px.Constant("Total"), 'Cargo', 'Nivel'], # Define a hierarquia das caixas
    values='Quantidade',
    color='Cargo',
    title='Estrutura da Equipe: Cargo e Nível de Senioridade (Clique nas caixas!)'
)

fig_tree.show()
```

## **Exportando para html**

```python
fig_tree.write_html("meu_dashboard_interativo.html")
print("✅ Gráfico exportado com sucesso! Você pode baixar o .html e abrir no seu navegador.")
```
